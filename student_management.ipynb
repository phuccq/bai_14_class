```python
from difflib import get_close_matches

class Student:
    def __init__(self,stu_id,name,age):
        self.stu_id = stu_id
        self.name = name
        self.age = age
    
class Student_List:
    def __init__(self):
        self.students = []
    def main(self):
        print('======= STUDENT MANAGEMENT PROGRAM =======')
        print('1. Student Information\n2. Add student\n3. Update student\n4. Delete student\n5. Lookup student\n6. Sort\n7. Exit')
        while True:
            user_choice = input('Select function: ')            
            if user_choice == '1':
                print('======= STUDENT INFORMATION =======')
                self.info(self.students)
            elif user_choice == '2':
                print('======= ADD STUDENT =======')
                self.add()
            elif user_choice == '3':
                print('======= UPDATE STUDENT =======')
                self.update()
            elif user_choice == '4':
                print('======= DELETE STUDENT =======')
                self.delete()
            elif user_choice == '5':
                print('======= LOOKUP STUDENT =======')
                self.lookup()
            elif user_choice == '6':
                print('======= SORT STUDENT LIST =======')
                self.sort_student()
            elif user_choice == '7':
                print('======= GOODBYE =======')
                break
            else:
                print("Ivalid function!")
                
    def info(self,students):
        print('ID'.ljust(5),'Name'.ljust(15),'age'.ljust(10))        
        for student in students:
            print(student.stu_id.ljust(5), student.name.ljust(15), student.age.ljust(10))           
            
    def add(self):
        run = True
        while run:
            new_stu_id = input('Input student\'s id: ')
            if self.students == []:
                run = False                
            else:
                for student in self.students:
                    if student.stu_id == new_stu_id:
                        print('Student\'s id is existed!')
                    else:
                        run = False     
        name = input('Input student\'s name: ')
        age = input('Input student\'s age: ')
        student = Student(new_stu_id,name,age)
        self.students.append(student)
        print('Add successful!')
        pass
    
    def find_student(self,student_info):
        result = []
        for student in self.students:
            if student.stu_id == student_info or student.name == student_info or student.age == student_info:
                result.append(student_info)
                if result != None:
                    return [result, student]
                else:
                    print('Student not founded!')
    
    def update(self):
        student_id = input('Input student\'s id to update: ')        
        if self.find_student(student_id) != None:
            result, student = self.find_student(student_id)            
            new_name = input('Update student\'s name: ')
            new_age = input('Update student\'s age: ')
            student.name = new_name
            student.age = new_age
            print('Update successful!')  
    
    def delete(self):
        student_id = input('Input student\'s id to delete: ')        
        if self.find_student(student_id) != None:
            result, student = self.find_student(student_id)
            self.students.pop(self.students.index(student))
            print('Delete successful!') 
                
    def lookup(self):
        while True:
            print('1. Lookup by id\n2. Lookup by name\n3. Lookup by age')
            option = input('Select option :')
            if option in ('1','2','3'):
                break
            else:
                print("Ivalid option!")
                
        def lookup_by(lookup_info):
            if self.find_student(lookup_info) != None:
                result, student = self.find_student(lookup_info)
                print(student.stu_id.ljust(5), student.name.ljust(15), student.age.ljust(10))        
                            
        if option == '1':
            lookup_id = input('Input student\'s id: ')
            print('ID'.ljust(5),'Name'.ljust(15),'age'.ljust(10))              
            lookup_by(lookup_id)
        elif option == '2':
            lookup_name = input('Input student\'s name: ')
            print('ID'.ljust(5),'Name'.ljust(15),'age'.ljust(10))              
            lookup_by(lookup_name)
        elif option == '3':
            lookup_age = input('Input student\'s age: ')
            print('ID'.ljust(5),'Name'.ljust(15),'age'.ljust(10))              
            lookup_by(lookup_age) 
            
    def sort_student(self):
        while True:
            print('1. Sort by id\n2. Sort by name\n3. Sort by age')
            option = input('Select option :')
            if option in ('1','2','3'):
                break
            else:
                print("Ivalid option!")
        def sort_by(lookup_info):
            sorted_students = self.students.copy()
            info_list = []
            def myFunc(student):
                if lookup_info == 'stu_id':
                    return student.stu_id
                if lookup_info == 'name':
                    return student.name
                if lookup_info == 'age':
                    return student.age
            sorted_students.sort(key=myFunc)
            self.info(sorted_students)
                
        if option == '1':
            sort_by('stu_id')
        elif option == '2':
            sort_by('name')            
        elif option == '3':
            sort_by('age')           


student = Student_List()
student.main()
    
```

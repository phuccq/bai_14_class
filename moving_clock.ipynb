```python
import turtle
import datetime

turtle.tracer(0,0)

screen = turtle.Screen()
screen.bgcolor('blue')


class Clock:    
    def __init__(self):
        # Khởi tạo thuộc tính vẽ mặt đồng hồ
        self.clock_surface = turtle.Turtle() 
        self.clock_surface.hideturtle()
        # Khởi tạo thuộc tính vẽ kim đồng hồ
        self.clock_hands = turtle.Turtle()        
        self.clock_hands.hideturtle()
        # Khởi tại thuộc tính thời gian
        self.clock_time()
        
    def clock_time(self):
        self.hour = datetime.datetime.now().hour
        self.minute = datetime.datetime.now().minute        
        self.second = datetime.datetime.now().second
                
    def draw_clock_surface(self):
        self.clock_surface.fillcolor('white')
        self.clock_surface.begin_fill()
        self.clock_surface.speed(0)
        self.clock_surface.penup()
        self.clock_surface.goto(0,-100)
        self.clock_surface.pendown()
        self.clock_surface.circle(100)
        self.clock_surface.end_fill()
        
        self.clock_surface.penup()
        self.clock_surface.goto(0,0)  
        self.clock_surface.dot(5)
        self.clock_surface.right(90)                
        for i in range(60):
            self.clock_surface.goto(0,0)  
            self.clock_surface.forward(90)
            self.clock_surface.pendown()
            if i%5 == 0:
                self.clock_surface.backward(10)
            else:
                self.clock_surface.backward(5)                
            self.clock_surface.left(6)
            self.clock_surface.penup()
        
    def draw_clock_hands(self):
        self.clock_hands.clear()
        hour_hand_move = ((self.hour % 12) * 3600 + self.minute * 60 + self.second) / (12*3600) * 360
        minute_hand_move = (self.minute * 60 + self.second) / 3600 * 360
        second_hand_move = (self.second) / 60 * 360
        
        #Vẽ kim giờ        
        self.clock_hands.pencolor('black')
        self.clock_hands.penup()
        self.clock_hands.goto(0,0)
        self.clock_hands.setheading(90)
        self.clock_hands.right(hour_hand_move)
        self.clock_hands.pendown()
        self.clock_hands.pensize(4)
        self.clock_hands.forward(50)  
        # Vẽ kim phút
        self.clock_hands.pencolor('black')
        self.clock_hands.penup()
        self.clock_hands.goto(0,0)
        self.clock_hands.setheading(90)        
        self.clock_hands.right(minute_hand_move)
        self.clock_hands.pendown()
        self.clock_hands.pensize(2)
        self.clock_hands.forward(60)
        # Vẽ kim giây
        self.clock_hands.pencolor('red')
        self.clock_hands.penup()
        self.clock_hands.goto(0,0)
        self.clock_hands.setheading(90)        
        self.clock_hands.right(second_hand_move)
        self.clock_hands.pendown()
        self.clock_hands.forward(75)

    def run(self):
        self.draw_clock_surface()
        self.moving_run()
        
    def moving_run(self): 
        self.clock_time()
        self.draw_clock_hands()        
        turtle.update()
        turtle.ontimer(self.moving_run, 1000)   

clock = Clock()
clock.run()

turtle.done()
```
